```python
#Cratd By Ameer Hamza
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the World Happiness Report dataset (adjust the file path)
import pandas as pd


# Load dataset 1 (CSV)
df1 = pd.read_csv('2015.csv')

# Load dataset 2
df2 = pd.read_csv('2016.csv')

# Load dataset 3 
df3 = pd.read_csv('2017.csv')

# Load dataset 4 .
df4 = pd.read_csv('2018.csv', delimiter='\t')

# Now you have four separate DataFrames: df1, df2, df3, and df4, each containing the data from its respective dataset.


```

```python
# Add a 'Year' column to each DataFrame to keep track of the year
df1['Year'] = 2015
df2['Year'] = 2016
df3['Year'] = 2017
df4['Year'] = 2018

# Concatenate the DataFrames vertically
combined_df = pd.concat([df1, df2, df3, df4], ignore_index=True)
```

```python
# Trends in Happiness Score Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Happiness Score', data=combined_df, ci=None)
plt.title('Trends in Happiness Score Over Time')
plt.xlabel('Year')
plt.ylabel('Happiness Score')
plt.show()
```

```python
# Regional Analysis Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Happiness Score', hue='Region', data=combined_df)
plt.title('Regional Trends in Happiness Score Over Time')
plt.xlabel('Year')
plt.ylabel('Happiness Score')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# correlation Analysis
correlation_matrix = combined_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Distribution of Happiness Score:
plt.figure(figsize=(10, 6))
sns.histplot(combined_df['Happiness Score'], bins=20, kde=True)
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()
```

```python
#Comparing Happiness Scores by Region
plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y='Happiness Score', data=combined_df, palette='Set1')
plt.xticks(rotation=90)
plt.title('Happiness Scores by Region')
plt.xlabel('Region')
plt.ylabel('Happiness Score')
plt.show()
```

```python
# Top 10 happiest countries
top_10_happiest = combined_df.nlargest(10, 'Happiness Score')

# Bottom 10 least happy countries
bottom_10_least_happy = combined_df.nsmallest(10, 'Happiness Score')

# Combine both and create a bar chart
top_bottom_happiness = pd.concat([top_10_happiest, bottom_10_least_happy])
plt.figure(figsize=(12, 6))
sns.barplot(x='Happiness Score', y='Country', data=top_bottom_happiness, palette='viridis', hue='Year')
plt.title('Top and Bottom Countries by Happiness Score')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.legend(title='Year')
plt.show()
```

```python
#Impact of GDP per Capita on Happiness
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Economy (GDP per Capita)', y='Happiness Score', data=combined_df, hue='Year', palette='Set2')
plt.title('Impact of GDP per Capita on Happiness Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
```

```python
#Average Happiness Score by Region
region_avg_happiness = combined_df.groupby(['Year', 'Region'])['Happiness Score'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Happiness Score', hue='Region', data=region_avg_happiness)
plt.title('Average Happiness Score by Region Over Time')
plt.xlabel('Year')
plt.ylabel('Average Happiness Score')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
